```python
# Module 1 Lab: From Data to Decision

import pandas as pd
from IPython.display import display

# Step 1: Create data
student_data = pd.DataFrame({
    "student_id": [1, 2, 3, 4, 5],
    "logins": [10, 2, 5, 1, 8],
    "assignment_submitted": [1, 0, 1, 0, 1],
    "forum_posts": [5, 0, 2, 0, 3]
})

print("Student engagement data:")
display(student_data)

# Step 2: Summarise data
print("Summary statistics:")
display(student_data.describe().round(2))

# Step 3: Apply a simple risk rule
student_data["at_risk"] = (
    (student_data["logins"] < 3) |
    (student_data["assignment_submitted"] == 0) |
    (student_data["forum_posts"] == 0)
)

print("Student engagement data with risk flag:")
display(student_data)
```
