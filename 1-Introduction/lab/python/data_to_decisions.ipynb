## Step 1: Load the data

Each row represents one student.

Look at the data first before making any judgment.

```python
df = pd.DataFrame({
    "Student": ["A", "B", "C", "D"],
    "Logins": [10, 2, 5, 0],
    "Assignment_Submitted": ["Yes", "No", "Yes", "No"],
    "Forum_Activity": ["High", "Low", "Low", "None"]
})

display(df)
```

## Step 3: Define a simple risk rule

We will now apply a simple rule-based interpretation.

This is not a predictive model.

It is a basic way to turn data into a decision.

```python
def assess_risk(row):
    if row["Logins"] == 0 and row["Assignment_Submitted"] == "No":
        return "High Risk"
    elif row["Logins"] <= 2:
        return "Medium Risk"
    else:
        return "Low Risk"

df["Risk_Level"] = df.apply(assess_risk, axis=1)

display(df)
```

## Guided interpretation

Look at the assigned risk levels and think about:

- Do you agree with all of these labels?

Write your observations in a new markdown cell.

## Step 4: Recommend actions

Once we identify risk, we need to decide what to do.

We will map each risk level to a simple action.

```python
def recommend_action(risk):
    if risk == "High Risk":
        return "Immediate intervention"
    elif risk == "Medium Risk":
        return "Monitor closely"
    else:
        return "No action needed"

df["Recommended_Action"] = df["Risk_Level"].apply(recommend_action)

display(df)
```
