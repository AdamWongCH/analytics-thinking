```python
import pandas as pd

# Sample dataset
data = {
    "student_id": [1, 2, 3, 4, 5],
    "logins": [10, 2, 5, 1, 8],
    "assignment_submitted": [1, 0, 1, 0, 1],
    "forum_posts": [5, 0, 2, 0, 3]
}

df = pd.DataFrame(data)

# Step 1: View data
print(df)

# Step 2: Basic summary
print(df.describe())

# Step 3: Simple rule-based risk flag
df["at_risk"] = (
    (df["logins"] < 3) |
    (df["assignment_submitted"] == 0) |
    (df["forum_posts"] == 0)
)

print(df)
```
