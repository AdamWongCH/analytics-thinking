```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

## Step 1: Load the data

We begin with a simple dataset.

Before building a model, first look at the values and ask:

- Do the variables make sense?
- Do higher engagement values seem to come with higher performance?
- Are there any unusual cases?

```python
student_data = pd.DataFrame({
    "student_id": [1, 2, 3, 4, 5, 6, 7, 8],
    "logins": [2, 4, 5, 7, 8, 10, 12, 15],
    "forum_posts": [0, 1, 1, 2, 3, 4, 5, 6],
    "performance_score": [45, 50, 55, 60, 66, 72, 78, 85]
})

display(student_data)
```

## Step 4: Visualise relationships

Before modelling, it helps to *see* the relationship.

We will plot:

- logins vs performance
- forum_posts vs performance

```python
plt.scatter(student_data["logins"], student_data["performance_score"])
plt.xlabel("Logins")
plt.ylabel("Performance Score")
plt.title("Logins vs Performance")
plt.show()
```

### Guided interpretation

## Step 5: Define the modelling problem

We want to predict:

- Target: `performance_score`

Using:

- Inputs: `logins`, `forum_posts`

```python
X = student_data[["logins", "forum_posts"]]
y = student_data["performance_score"]
```

## Step 6: Build the regression model

We will now fit a simple linear regression model.

The model will try to learn the relationship between:

- engagement variables
- student performance

```python
model = LinearRegression()
model.fit(X, y)
```

## Step 7: Interpret coefficients

The coefficients tell us the direction and strength of relationships.

Think about:

- If a coefficient is positive, what does that mean?
- If a coefficient is larger, what might that suggest?

```python
coefficients = pd.DataFrame({
    "variable": X.columns,
    "coefficient": model.coef_
})

display(coefficients)
print("Intercept:", round(model.intercept_, 2))
```

### Guided interpretation

- Are the relationships positive or negative?
- Which variable has a stronger effect?
- Do the directions of the relationships make sense?

Write a short interpretation in a new markdown cell.

## Step 8: Generate predictions

Now let us use the model to predict performance scores for the same dataset.

This is not yet a full evaluation, but it helps us see whether the model roughly follows the pattern in the data.

```python
student_data["predicted_score"] = model.predict(X)
display(student_data.round(2))
```

## Step 9: Evaluate the model

We will use two simple metrics:

- **Mean Absolute Error (MAE)**: how far predictions are from actual scores on average
- **R²**: how much of the variation in performance is explained by the model

```python
mae = mean_absolute_error(y, student_data["predicted_score"])
r2 = r2_score(y, student_data["predicted_score"])

print("MAE:", round(mae, 2))
print("R²:", round(r2, 2))
```
