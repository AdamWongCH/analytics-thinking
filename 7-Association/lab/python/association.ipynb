## Step 3: Define simple rules

We will examine rules of the form:

- `A -> B`

This means:

> when A occurs, how often does B also occur?

We will calculate three measures:

- **Support**: how often A and B occur together
- **Confidence**: when A happens, how often does B happen?
- **Lift**: how much more likely B is when A happens?

```python
def support(df, antecedent, consequent):
    both = ((df[antecedent] == 1) & (df[consequent] == 1)).sum()
    return both / len(df)

def confidence(df, antecedent, consequent):
    antecedent_count = (df[antecedent] == 1).sum()
    both = ((df[antecedent] == 1) & (df[consequent] == 1)).sum()
    return both / antecedent_count if antecedent_count > 0 else 0

def lift(df, antecedent, consequent):
    conf = confidence(df, antecedent, consequent)
    consequent_support = (df[consequent] == 1).sum() / len(df)
    return conf / consequent_support if consequent_support > 0 else 0
```

## Step 4: Generate a sample rule

Let us examine this rule:

`high_logins -> forum_participation`

This asks:

> among students with high logins, how often do they also participate in forums?

```python
rule_support = support(student_data, "high_logins", "forum_participation")
rule_confidence = confidence(student_data, "high_logins", "forum_participation")
rule_lift = lift(student_data, "high_logins", "forum_participation")

print("Support:", round(rule_support, 2))
print("Confidence:", round(rule_confidence, 2))
print("Lift:", round(rule_lift, 2))
```

## Guided interpretation

Interpret the results:

## Step 5: Compare multiple rules

Now let us compare several possible rules.

```python
rules = [
    ("high_logins", "forum_participation"),
    ("high_logins", "assignment_submitted"),
    ("forum_participation", "assignment_submitted")
]

results = []

for antecedent, consequent in rules:
    results.append({
        "Rule": f"{antecedent} -> {consequent}",
        "Support": round(support(student_data, antecedent, consequent), 2),
        "Confidence": round(confidence(student_data, antecedent, consequent), 2),
        "Lift": round(lift(student_data, antecedent, consequent), 2)
    })

rules_df = pd.DataFrame(results)
display(rules_df)
```
