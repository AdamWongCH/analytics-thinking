## Scenario

You are given student engagement data.

Your task is to identify different types of students based on their engagement.

There are no labels.

You must discover the groups.

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.cluster import KMeans
```

## Step 1: Load the data

Observe the dataset.

Look for:
- patterns
- variation
- possible groupings

```python
student_data = pd.DataFrame({
    "student_id": range(1, 13),
    "logins": [15,14,13, 8,7,6, 3,2,1, 10,9,4],
    "forum_posts": [7,6,6, 4,3,3, 1,0,0, 5,4,1]
})

display(student_data)
```

## Step 2: Visualise the data

Before clustering, plot the data.

This helps you see whether groups already exist.

We will plot:

- logins vs forum_posts

```python
plt.scatter(student_data["logins"], student_data["forum_posts"])
plt.xlabel("Logins")
plt.ylabel("Forum Posts")
plt.title("Student Engagement")
plt.show()
```

## Guided interpretation

Look at the plot:

- Do you see natural groupings?
- Are there clusters forming?
- How many groups might exist?

Write your observations in a new markdown cell.

## Step 3: Apply clustering

We will use K-Means clustering.

You need to choose the number of clusters (k).

We will start with k = 3.

```python
kmeans = KMeans(n_clusters=3, random_state=42)
student_data["cluster"] = kmeans.fit_predict(student_data[["logins", "forum_posts"]])

display(student_data)
```

## Step 4: Visualise clusters

Plot the data again, but now colour by cluster.

This helps you see how the algorithm grouped the data.

```python
plt.scatter(
    student_data["logins"],
    student_data["forum_posts"],
    c=student_data["cluster"]
)

plt.xlabel("Logins")
plt.ylabel("Forum Posts")
plt.title("Clusters of Student Engagement")
plt.show()
```

## Guided interpretation

Look at the clusters:

- Do the groups make sense?
- Are the clusters clearly separated?
- What defines each cluster?

Write your observations in a new markdown cell.

## Step 5: Understand cluster characteristics

Let us summarise each cluster.

This helps us interpret what each group represents.

```python
cluster_summary = student_data.groupby("cluster").mean(numeric_only=True)
display(cluster_summary.round(2))
```
