## Scenario

You are given weekly student engagement data.

Your task is to identify patterns over time, interpret what they might mean, and make a simple forecast.

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

## Step 1: Load the data

We begin with a simple weekly dataset.

Look at the data and ask:

- What does each row represent?
- What is the time variable?
- What is being measured over time?

```python
engagement_data = pd.DataFrame({
    "week": [1, 2, 3, 4, 5, 6, 7, 8],
    "logins": [120, 118, 125, 130, 128, 115, 98, 90]
})

display(engagement_data)
```

Before plotting the data, understand:

## Step 3: Visualise the data

The best way to understand time series is to plot it.

We will create a line chart of logins over time.

```python
plt.plot(engagement_data["week"], engagement_data["logins"], marker="o")
plt.xlabel("Week")
plt.ylabel("Logins")
plt.title("Weekly Student Logins")
plt.show()
```

## Step 4: Identify trend

A trend is the general direction of movement over time.

We can compare the start and end of the series to get a simple sense of trend.

```python
start_value = engagement_data["logins"].iloc[0]
end_value = engagement_data["logins"].iloc[-1]

print("Start value:", start_value)
print("End value:", end_value)
print("Change:", end_value - start_value)
```

## Guided interpretation

Reflect on the change:

- Is the overall trend upward or downward?

Write a short interpretation in a new markdown cell.

## Step 5: Smooth the series with a moving average

Weekly values can fluctuate.

A moving average helps us see the underlying pattern more clearly by smoothing short-term variation.

We will use a 3-week moving average.

```python
engagement_data["moving_average_3"] = engagement_data["logins"].rolling(window=3).mean()
display(engagement_data)
```

## Step 6: Plot the original data and moving average

This helps us compare:

- actual weekly values
- the smoothed trend

```python
plt.plot(engagement_data["week"], engagement_data["logins"], marker="o", label="Actual Logins")
plt.plot(engagement_data["week"], engagement_data["moving_average_3"], marker="o", label="3-Week Moving Average")
plt.xlabel("Week")
plt.ylabel("Logins")
plt.title("Weekly Logins with Moving Average")
plt.legend()
plt.show()
```

## Guided interpretation

Look at the smoothed line and think about:

- Does the moving average make the pattern clearer?
- Does it confirm the same trend?
- Is there a period where engagement seems to decline more clearly?

## Step 8: A simple forecast

So far, we have:

- observed past behaviour
- identified trends
- smoothed the data

Now we ask:

> What might happen next?

We will create a simple forecast using the recent trend.

This is not meant to be precise — it is meant to provide a rough indication of future behaviour.

```python
last_ma = engagement_data["moving_average_3"].iloc[-1]

print("Last moving average:", round(last_ma, 2))

forecast_week = 9
forecast_value = last_ma

print(f"Forecast for Week {forecast_week}:", round(forecast_value, 2))
```

## Guided interpretation

This forecast assumes:

- the recent pattern continues
- short-term fluctuations are smoothed out

Think about:

- Is this a reasonable assumption?
- What could cause the actual value to differ?

Write a short interpretation in a new markdown cell.

```python
forecast_df = pd.DataFrame({
    "week": [forecast_week],
    "logins": [forecast_value]
})

combined = pd.concat([engagement_data[["week", "logins"]], forecast_df], ignore_index=True)

plt.plot(engagement_data["week"], engagement_data["logins"], marker="o", label="Actual")
plt.plot(engagement_data["week"], engagement_data["moving_average_3"], label="Moving Average")
plt.scatter(forecast_week, forecast_value, color="red", label="Forecast")

plt.xlabel("Week")
plt.ylabel("Logins")
plt.title("Forecasted Engagement")
plt.legend()
plt.show()
```

## Step 10: Limitations of simple forecasting

This simple forecast has limitations:

- it only uses recent values
- it does not account for sudden changes
- it assumes patterns continue

In reality, forecasting often requires:

- more data

- understanding of context
