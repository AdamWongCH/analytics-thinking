```python
import pandas as pd
import re
from collections import Counter
from IPython.display import display
```

## Step 1: Load the data

We begin with a small set of student comments.

Read them first before doing any analysis.

Ask:

- What issues seem to appear?
- What words stand out?
- Are the comments mostly positive, negative, or mixed?

```python
feedback_data = pd.DataFrame({
    "comment": [
        "The course content is interesting but the workload is too heavy.",
        "The instructor is helpful and explains concepts clearly.",
        "Too many assignments and not enough time to complete them.",
        "I enjoyed the lessons, but the pace was too fast.",
        "The examples were useful and practical.",
        "The instructions for the assignment were unclear.",
        "The course is engaging and well-structured.",
        "I found the workload stressful and difficult to manage."
    ]
})

display(feedback_data)
```

## Step 2: Clean the text

Before analysing text, we need to clean it.

We will:

- convert text to lowercase
- remove punctuation
- split text into words

This helps standardise the text.

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    return text

feedback_data["clean_comment"] = feedback_data["comment"].apply(clean_text)
display(feedback_data)
```

## Step 3: Tokenise the text

Tokenising means splitting text into individual words.

This allows us to count and analyse words.

```python
feedback_data["tokens"] = feedback_data["clean_comment"].apply(str.split)
display(feedback_data[["comment", "tokens"]])
```

## Step 4: Count word frequencies

We now combine all words and count how often they appear.

This helps us identify the most common terms in the feedback.

```python
all_words = [word for tokens in feedback_data["tokens"] for word in tokens]
word_counts = Counter(all_words)

word_freq = pd.DataFrame(word_counts.items(), columns=["word", "count"]).sort_values(
    by="count", ascending=False
)

display(word_freq.head(15))
```

## Step 5: Remove common stopwords

Some words are common but not very meaningful for interpretation.

Examples include:

- the
- is
- and
- to

We remove these to make important words stand out more clearly.

```python
stopwords = {
    "the", "is", "and", "to", "but", "i", "a", "of", "for", "was", "too",
    "not", "them", "were", "this", "it", "in", "on", "my"
}

filtered_words = [word for word in all_words if word not in stopwords]
filtered_counts = Counter(filtered_words)

filtered_freq = pd.DataFrame(filtered_counts.items(), columns=["word", "count"]).sort_values(
    by="count", ascending=False
)

display(filtered_freq.head(15))
```

## Step 6: Perform simple sentiment analysis

We will use a very simple approach.

We define:

- a small list of positive words
- a small list of negative words

Then we count how many positive and negative words appear in each comment.

This is only a basic method, but it helps illustrate the idea.

```python
positive_words = {"interesting", "helpful", "clearly", "enjoyed", "useful", "practical", "engaging", "wellstructured"}
negative_words = {"heavy", "stressful", "difficult", "unclear", "fast"}

def sentiment_score(tokens):
    score = 0
    for word in tokens:
        if word in positive_words:
            score += 1
        elif word in negative_words:
            score -= 1
    return score

feedback_data["sentiment_score"] = feedback_data["tokens"].apply(sentiment_score)
display(feedback_data[["comment", "sentiment_score"]])
```

## Step 7: Label simple sentiment

We now convert the sentiment score into a simple label:

- positive
- neutral
- negative

```python
def sentiment_label(score):
    if score > 0:
        return "positive"
    elif score < 0:
        return "negative"
    else:
        return "neutral"

feedback_data["sentiment_label"] = feedback_data["sentiment_score"].apply(sentiment_label)
display(feedback_data[["comment", "sentiment_score", "sentiment_label"]])
```
